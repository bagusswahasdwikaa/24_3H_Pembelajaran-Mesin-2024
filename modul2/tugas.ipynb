```python
#1. Pisahkan antara variabel yang dapat digunakan dan variabel yang tidak dapat digunakan
import pandas as pd 

data = 'wbc.csv' # path dataset
df = pd.read_csv(data) #load dataset

df.head()
```

```python
df.info()
```

```python
#BERDASARKAN INFORMASI DARI DATA DIATAS VARIABEL ID, DIAGNOSIS TIDAK TERMASUK PADA VARIABEL YANG DIGUNAKAN DAN JUGA SAYA MENGABAIKAN VARIABEL YANG TIDAK MEMILIKI NILAI PADA KOLOM KE-33
# Identifikasi kolom ke-33 tidak memiliki nilai
col_33 = df.columns[32] 

# Mengidentifikasi variabel yang digunakan
used_columns = df.select_dtypes(include=['float64']).columns

# Mengabaikan variabel yang seluruhnya memiliki nilai NaN
used_columns_clean = [col for col in used_columns if col != col_33 and not df[col].isnull().all()]
df_used = df[used_columns_clean]

# Mengidentifikasi variabel tidak digunakan
non_used_columns = df.select_dtypes(exclude=['float64']).columns
df_non_used = df[non_used_columns]

print("Variabel yang dapat digunakan (kolom numerik):")
print(df_used.head())

print("\nVariabel yang tidak dapat digunakan (kolom non-numerik):")
print(df_non_used.head())
```

```python
#2. Lakukan proses encoding pada kolom "diagnosis".
from sklearn.preprocessing import LabelEncoder, StandardScaler
le = LabelEncoder() # membuat objek dari LabelEncoder
df['diagnosis'] = le.fit_transform(df['diagnosis']) # proses encoding
df.head()
```

```python
#3. Lakukan proses standarisasi pada semua kolom yang memiliki nilai numerik.
from sklearn.preprocessing import StandardScaler

# Identifikasi kolom numerik
numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns


scaler = StandardScaler()
# Abaikan kolom numerik yang seluruhnya NaN
numerical_columns_clean = [col for col in numerical_columns if not df[col].isnull().all()]

# Terapkan standarisasi hanya pada kolom yang valid
df_numerical_clean = df[numerical_columns_clean]
df_numerical_standardized = df_numerical_clean.copy()

# Fitting dan transformasi data
df_numerical_standardized[numerical_columns_clean] = scaler.fit_transform(df_numerical_clean)

print("Data setelah standarisasi di semua kolom numerik:")
print(df_numerical_standardized.head())
print("Menampilkan semua kolom :")
df.head()
```

```python
#4. Lakukan proses stratified split data untuk membuat data latih dan data uji dengan rasio 80:20.
from sklearn.model_selection import train_test_split
#  Kolom 'diagnosis' adalah kolom target untuk stratifikasi
target_column = 'diagnosis'

# Melakukan stratified split dengan rasio 80:20
df_train, df_test = train_test_split(df, test_size=0.2, stratify=df[target_column], random_state=42)

# Menampilkan ukuran data latih dan data uji
print(f"Jumlah data asli: {len(df)}")
print(f"Jumlah data latih: {len(df_train)}")
print(f"Jumlah data uji: {len(df_test)}")
```
